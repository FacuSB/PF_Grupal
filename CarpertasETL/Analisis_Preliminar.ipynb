## 4_ Unicidad

Identifico y Elimino duplicados en los datos

#### Pruebas

funcion para contar registros duplicados

```python
def count_duplicate_rows(df):
    # Convierte los diccionarios y las listas a cadenas
    df = df.applymap(lambda x: str(x) if isinstance(x, (dict, list)) else x)
    
    # Cuenta las filas duplicadas
    duplicate_rows_df = df[df.duplicated()]
    
    # Devuelve el n√∫mero de filas duplicadas
    return len(duplicate_rows_df)
```

```python
count_duplicate_rows(yelp_business_df)
```

```python
count_duplicate_rows(yelp_checkin_df)
```

```python
count_duplicate_rows(yelp_review_df)
```

```python
count_duplicate_rows(yelp_tip_df)
```

```python
count_duplicate_rows(yelp_user_df)
```

```python
count_duplicate_rows(google_metadata_1)
```

```python
count_duplicate_rows(google_reviews_florida_1)
```

#### Conclusion

hay bastantes registros duplicados, esto se va a tener que limpiar con una funcion para sacar los registros duplicados de todas los datasets que se van a utilizar mas adelante
